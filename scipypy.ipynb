```python
import numpy as np
from matplotlib.colors import ListedColormap
N=500
current_palette=sns.color_palette("muted",n_colors=5)
cmap=ListedColormap(sns.color_palette(current_palette).as_hex())
data1=np.random.randn(N)
data2=np.random.randn(N)
colors=np.random.randint(0,5,N)
plt.scatter(data1,data2,c=colors,cmap=cmap)
plt.colorbar()
plt.show()
```

```python
from scipy import misc
face=misc.face(gray=True)
face=face[:512,-512:]
from scipy import ndimage
from scipy import signal
from matplotlib import pyplot as ply
import numpy as np
noisy_face=np.copy(face).astype(np.float)
noisy_face+=face.std()*0.5*np.random.standard_normal(face.shape)
blurred_face=ndimage.gaussian_filter(noisy_face,sigma=5)
median_face=ndimage.median_filter(noisy_face,(5,5))
wiener_face=signal.wiener(noisy_face,(5,5))

plt.figure(figsize=(15,4))
plt.subplot(151)
plt.imshow(noisy_face, cmap=plt.cm.gray)
plt.axis('off')

plt.subplot(152)
plt.imshow(blurred_face, cmap=plt.cm.gray)
plt.axis('off')

plt.subplot(153)
plt.imshow(median_face, cmap=plt.cm.gray)
plt.axis('off')

plt.subplot(154)
plt.imshow(wiener_face, cmap=plt.cm.gray)
plt.axis('off')

plt.subplots_adjust(wspace=.05, left=.01, bottom=.01, right=0.99, top=.99)
plt.show()
```
